```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example DataFrame creation (you should load your actual DataFrame)
data = pd.DataFrame({
    'Date': pd.date_range(start='1/1/2020', periods=100, freq='M'),
    'Ward': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C', 'D'] * 10
})

# Print column names to verify
print(data.columns)

# Set figure size and style
plt.rcParams['figure.figsize'] = (20, 9)
plt.style.use('dark_background')

# Create the count plot
sns.countplot(x='Ward', data=data, palette='gnuplot')

# Set plot title and rotate x-axis labels
plt.title('Major Crimes in Chicago', fontweight='bold', fontsize=20)
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Check available styles
print(plt.style.available)

# Set a valid style
plt.style.use('seaborn-v0_8-pastel')

# Setting plot parameters
plt.rcParams['figure.figsize'] = (20, 9)

# Generating colors
color = plt.cm.spring(np.linspace(0, 1, 15))

# Assuming 'data' is your DataFrame
data['Location Description'].value_counts().plot.bar(color=color)

# Adding title and formatting x-axis labels
plt.title('District with Most Crime', fontsize=30)
plt.xticks(rotation=90)

# Displaying the plot
plt.show()
```

```python
# Regions with count of crimes

plt.rcParams['figure.figsize'] = (20, 9)
plt.style.use('seaborn-v0_8-pastel')

color = plt.cm.ocean(np.linspace(0, 1, 15))
data['Block'].value_counts().head(15).plot.bar(color = color, figsize = (15, 10))

plt.title('Top 15 Regions in Crime',fontsize = 20)

plt.xticks(rotation = 90)
plt.show()
```

```python
plt.style.use('seaborn-v0_8')

color = plt.cm.winter(np.linspace(0, 10, 20))
data['Primary Type'].value_counts().plot.bar(color = color, figsize = (15, 8))

plt.title('Resolutions for Crime',fontsize = 20)
plt.xticks(rotation = 90)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = pd.DataFrame({
    'Date': pd.date_range(start='1/1/2020', periods=100, freq='M')
})

# Convert 'Date' column to datetime
data['Date'] = pd.to_datetime(data['Date'])

# Extract month from the 'Date' column
data['Month'] = data['Date'].dt.month

# Set the plot style and figure size
plt.style.use('seaborn-v0_8-white')
plt.rcParams['figure.figsize'] = (15, 8)

# Create the count plot
sns.countplot(x='Month', data=data, palette='autumn')
plt.title('Crimes in each Month', fontsize=20)

# Show the plot
plt.show()
```
